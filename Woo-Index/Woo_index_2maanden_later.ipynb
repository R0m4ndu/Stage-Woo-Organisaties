# Analyse Woo/verwijs index 2 maanden later

Er zijn inmiddels al 447 bestuursorganen die tenminste een deel van de Woo index hebben inbgevuld. Met nog steeds 17 informatiecategorieën. Maar nu wel met 2014 links naar plaatsten op het web. Oorspronkelijk waren er gemiddeld 4.16 links per bestuursorgaan en dit zijn er nu 4.51. Wat toch aantoont dat de verwijsindexen in aantallen en in dichtheid groeien. 


```python
import json
import pandas as pd
import seaborn as sns
from collections import Counter
```

```python
wdf=pd.read_json('woo_index.json')
wdf['nr_of_indexes']=wdf.foi_files.apply(lambda x:len(x['foi_documenten']))
wdf['organisatietype']=wdf.dc_publisher.str.replace(r'\d','', regex=True)
S=wdf.foi_files.apply(lambda x:pd.DataFrame(x['foi_documenten']).dc_title.values)
D= pd.DataFrame([Counter(i) for i in S     ])
Woo_categorieen=sorted(D.columns)
wdf=wdf.join(D)
print(len(Woo_categorieen),Woo_categorieen)
wdf.head(2)
```

```python
print('Verdeling van het aantal ingevulde informatiecategorieen per bestuursorgaan')
wdf.nr_of_indexes.describe()
```

```python
print('Per organisatietype, de verdeling van het aantal ingevulde informatiecategorieen.')
wdf.groupby('organisatietype')['nr_of_indexes'].describe().round(0).astype(int)
```

```python
print('Per Woo-categorie aantal bestuursorganen die die categorie hebben ingevuld')
wdf[Woo_categorieen].sum().astype(int)
```

```python
print('Hetzelfde maar nu per organisatietype')
W=wdf.groupby('organisatietype')[Woo_categorieen].sum().T.astype(int)
W
```
