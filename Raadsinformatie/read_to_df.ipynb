```python
import pandas as pd
import gzip
import json
import os
from collections import Counter
```

```python
directory = 'alle_gemeente_rapporten'

all_categories = []

rapport_total = 0
doc_total = 0
total_pages = 0

for file in os.listdir(directory)[1:]:

    file_path = f"{directory}/{file}"
    
    with gzip.open(file_path, 'rt') as f:  # Use 'rt' for reading text files, 'rb' for binary files
        content = f.read()
    
    data_dict = json.loads(content)
    
    total = 0
    for i in data_dict:
        all_categories.append(i['foi_classificaties'])
        
        total += len(i['foi_files'])

        if len(i['foi_files']) != 0:
            for page in i['foi_files']:
                total_pages += len(page['foi_bodyText'])

    doc_total += total
    
    rapport_total += len(data_dict)

print(doc_total)
print(rapport_total)
print(total_pages)
```
