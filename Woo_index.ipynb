# Analyse Woo/verwijs index

De *verwijsindex* uit het BIT advies over PLOOI, vorig jaar ook beloofd door de minister komt nu rap van de grond, en is in te zien op organisaties.overheid.nl. Woogle heeft wat geteld en geturfd, en geeft hier een overzicht van de stand van zaken.

Er zijn 402 bestuursorganen die in ieder egvale en deel van de Woo-index hebben ingevuld. Er zijn 17 informatiecategorieen in te vullen. In totaal zijn er 1.672 links naar plaatsen op het web met Woo-informatie ingevoerd. Dit is een beetje een overtekeing want meer dan 1.000 links zijn al automatisch ingevuld, voor 3 categorieen, die al centraal gepubliceerd worden.

Maar goed, dit is een geweldig mooi begin. Op naar stap 2, de zoekfunctie. Daarvoor zal alle informatie van al die links opgehaald en geindexeerd moeten worden. Woogle doet dit al , *voorzover dat automatisch haalbaar is*.

```python
import json
import pandas as pd
import seaborn as sns
from collections import Counter
```

```python
wdf=pd.read_json('woo_index.json')
wdf['nr_of_indexes']=wdf.foi_files.apply(lambda x:len(x['foi_documenten']))
wdf['organisatietype']=wdf.dc_publisher.str.replace(r'\d','', regex=True)
S=wdf.foi_files.apply(lambda x:pd.DataFrame(x['foi_documenten']).dc_title.values)
D= pd.DataFrame([Counter(i) for i in S     ])
Woo_categorieen=sorted(D.columns)
wdf=wdf.join(D)
print(len(Woo_categorieen),Woo_categorieen)
wdf.head(2)
```

```python
print('Verdeling van het aantal ingevulde informatiecategorieen per bestuursorgaan')
wdf.nr_of_indexes.describe()
```

```python
print('Per organisatietype, de verdeling van het aantal ingevulde informatiecategorieen.')
wdf.groupby('organisatietype')['nr_of_indexes'].describe().round(0).astype(int)
```

```python
print('Per Woo-categorie aantal bestuursorganen die die categorie hebben ingevuld')
wdf[Woo_categorieen].sum().astype(int)
```

```python
print('Hetzelfde maar nu per organisatietype')
W=wdf.groupby('organisatietype')[Woo_categorieen].sum().T.astype(int)
W
```
